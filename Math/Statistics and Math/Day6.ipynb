```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("products.csv")
dataset.head()
```

```python
dataset.info()
```

```python
data_corr = dataset.select_dtypes(["float64","int64"]).corr()
data_corr
```

```python
data_cov = dataset.select_dtypes(["float64","int64"]).cov()
data_cov
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(data_corr,annot=True)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(data_cov,annot=True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
x = np.array([2, 4, 6, 8, 10])
y = np.array([1, 3, 4, 6, 8])

# Create DataFrame
df = pd.DataFrame({"X": x, "Y": y})

# Covariance
cov_matrix = np.cov(x, y)  # by default, gives sample covariance
print("Covariance Matrix:\n", cov_matrix)
print("Covariance (X,Y):", cov_matrix[0, 1])

# Correlation
corr_matrix = np.corrcoef(x, y)
print("Correlation Matrix:\n", corr_matrix)
print("Correlation (X,Y):", corr_matrix[0, 1])
```

```python
# Scatter plot with regression line
sns.lmplot(x="X", y="Y", data=df, height=5, aspect=1.2)
plt.title("Scatter Plot with Regression Line")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
# Sample space for a fair die roll (1 to 6)
die_rolls = np.array([1, 2, 3, 4, 5, 6])
pmf = np.ones_like(die_rolls) / len(die_rolls)  # Each value has 1/6 chance

# Plot PMF
plt.figure(figsize=(7, 4))
plt.bar(die_rolls, pmf, color="skyblue")
plt.title("PMF of Fair Die Roll")
plt.xlabel("Die Face")
plt.ylabel("Probability")
plt.show()
```

```python
# Create data from normal distribution
mu, sigma = 0, 1
x = np.linspace(-4, 4, 1000)
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(- (x - mu)**2 / (2 * sigma**2))

# Plot PDF
plt.figure(figsize=(8, 4))
plt.plot(x, pdf, color='blue')
plt.title("Normal Distribution PDF (=0, =1)")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

```python
# CDF from the same data
cdf = np.cumsum(pdf)
cdf = cdf / cdf[-1]  # Normalize to make final value 1

# Plot CDF
plt.figure(figsize=(8, 4))
plt.plot(x, cdf, color='green')
plt.title("Normal Distribution CDF")
plt.xlabel("X")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()
```

```python
# Generate random sample data
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot using seaborn
plt.figure(figsize=(8, 4))
sns.histplot(data, kde=True, bins=30, color="orange")
plt.title("Random Sample from Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Standard normal
z = np.random.normal(0, 1, 1000)
# Custom normal
custom = np.random.normal(5, 2, 1000)

# Compare with KDE plots
plt.figure(figsize=(8, 4))
sns.kdeplot(z, label="Standard Normal (=0, =1)", fill=True)
sns.kdeplot(custom, label="Custom Normal (=5, =2)", fill=True)
plt.title("Standard vs Custom Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x,y = make_moons(n_samples=250,noise=0.05)
```

```python
df = {'data1':x[:,0],'data2':x[:,1]}
```

```python
dataset =  pd.DataFrame(df)
dataset.head()
```

```python
sns.scatterplot(x='data1',y='data2',data=dataset)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.2,min_samples=3)
dataset['Predict Columns']=db.fit_predict(dataset)
```

```python
sns.scatterplot(x='data1',y='data2',data=dataset,hue='Predict Columns')
plt.show()
```

```python
import pandas as pd
import scipy.cluster.hierarchy as sc
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('iris.csv')
dataset.head(3)
```

```python
# sns.pairplot(data=dataset)
# plt.show()
```

```python
plt.Figure(figsize=(25,25))
linkage_matrix = sc.linkage(dataset, method='single', metric='euclidean')
dendrogram = sc.dendrogram(linkage_matrix)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2,linkage='single')
dataset['predict'] = ac.fit_predict(dataset)
```

```python
dataset.head()
```

```python
sns.pairplot(data=dataset,hue='predict')
plt.show()
```
